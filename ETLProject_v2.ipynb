```python
genre_df=pd.read_csv('genre.csv')
genre_df
```

```python
genre_df=genre_df.rename(columns={"Titles":"Name"})
```

```python
book_genre_df=pd.merge(books_df, genre_df, on="Name")
book_genre_df
```

```python
book_genre_df=book_genre_df.rename(columns={"Genre_x":"Gen_Genre", "Genre_y":"Spec_Genre"})
book_genre_df
```

```python
book_genre_df = book_genre_df.drop(columns=["Unnamed: 0"])
book_genre_df
```

```python
race_df = book_genre_df[book_genre_df["Spec_Genre"]=="Race"]
race_df
```

```python
social_justice_df = book_genre_df[book_genre_df["Spec_Genre"]=="Social Justice"]
social_justice_df
```

```python
genre_count = book_genre_df.groupby(book_genre_df["Spec_Genre"]).count()
genre_count = genre_count.sort_values(by=["Name"],ascending=False)
genre_count.head(20)
```
